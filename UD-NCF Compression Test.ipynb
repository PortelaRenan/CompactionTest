plt.rcParams["figure.figsize"] = [7.50, 3.50]

fig, ax1 = plt.subplots()
ax1.grid()
ax2 = ax1.twinx()

x_ = np.arange(0.16, 900, 0.01)

columns1 = ['y1_' + str(i) for i in range(len(CF_0_45_4_20cst_AT))] + ['Time (s)']
df_force = pd.DataFrame(columns = columns1)

columns2 = ['y2_' + str(i) for i in range(len(CF_0_45_4_20cst_AT))] + ['Time (s)']
df_displacement = pd.DataFrame(columns = columns2)

df_force['Time (s)'] = df_displacement['Time (s)'] = x_

for i in range(len(CF_0_45_4_20cst_AT)):
    x = CF_0_45_4_20cst_AT[i].iloc[:,0].to_numpy()
    y1 = CF_0_45_4_20cst_AT[i].iloc[:,1].to_numpy()
    y2 = 6 - CF_0_45_4_20cst_AT[i].iloc[:,2].to_numpy()

    interp_func_1 = interp1d(x, y1)
    interp_func_2 = interp1d(x, y2)

    df_force['y1_' + str(i)] = interp_func_1(x_)
    df_displacement['y2_' + str(i)] = interp_func_2(x_)
    
ax1.plot(x_, df_force.iloc[:, 0:len(CF_0_45_4_20cst_AT)].mean(axis = 1).to_numpy(), color = 'red')
ax1.fill_between(x_, df_force.iloc[:, 0:len(CF_0_45_4_20cst_AT)].mean(axis = 1).to_numpy() - df_force.iloc[:, 0:len(CF_0_45_4_20cst_AT)].std(axis = 1).to_numpy(), 
                     df_force.iloc[:, 0:len(CF_0_45_4_20cst_AT)].mean(axis = 1).to_numpy() + df_force.iloc[:, 0:len(CF_0_45_4_20cst_AT)].std(axis = 1).to_numpy(),
